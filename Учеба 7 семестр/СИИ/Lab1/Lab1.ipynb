```python
import numpy as np
import pandas as pd 
from mpl_toolkits.mplot3d import axes3d
import matplotlib.pyplot as plt
import random 

#,   
def funct_x1(x):
    return x ** 2 - 50
def funct_x2(x):
    return 0.001 * x

# 
array_x = np.arange(1, 400, 1)
x1 = [funct_x1(x) for x in array_x]
x2 = [funct_x2(x) for x in array_x]
y = [random.randint(1, 400) for x in range(1, 400)]

#   numpy
x1 = np.array(x1)
x2 = np.array(x2)
y = np.array(y)

#  CSV   frame pandas`a 
df = pd.DataFrame({'x1':x1, 'x2':x2, 'y':y})
df.to_csv("Lab1_res.csv",sep= '\t', index = False)
print(df)

#  x1(x), x2(x) 
plt.plot(array_x, df[['x1']], label='x1')
plt.plot(array_x, df[['x2']], label='x2')
plt.grid(True)
plt.legend(loc='best')
plt.xlabel('x', fontsize=15)
plt.ylabel('x1, x2', fontsize=15)
plt.show()

#  y(x1), y(x2) 
plt.plot(df[['x1']], df[['y']], 'o', label='y(x1)')
plt.plot(df[['x2']], df[['y']], 'o', label='y(x2)')
plt.grid(True)
plt.legend(loc='best')
plt.xlabel('x1, x2', fontsize=15)
plt.ylabel('y', fontsize=15)
plt.show()

#   
print '   1: {0}' .format(np.mean(x1))
print '   2: {0}' .format(np.mean(x2))
print '   y: {0}\n' .format(np.mean(y))

#   
print '  1: {0}' .format(np.min(x1))
print '  2: {0}' .format(np.min(x2))
print '  y: {0}\n' .format(np.min(y))

#   
print '  1: {0}' .format(np.max(x1))
print '  2: {0}' .format(np.max(x2))
print '  y: {0}\n' .format(np.max(y))

#        
df2 = df[(df.x1 < np.mean(x1)) | (df.x2 < np.mean(x2))]
df.to_csv("Lab1_res_2.csv",sep= '\t', index = False)

#3D   y(x1, x2)    mplot3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
x1 = df[['x1']]
x2 = df[['x2']]
y = df[['y']]
ax.scatter(x1, x2, y)
plt.xlabel('x1', fontsize=15)
plt.ylabel('x2', fontsize=15)
plt.show()
```
