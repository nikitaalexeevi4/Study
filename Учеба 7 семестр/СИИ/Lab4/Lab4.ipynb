```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.metrics import f1_score
from sklearn.utils import shuffle
from sklearn.model_selection import train_test_split
from sklearn import preprocessing
from sklearn.cluster import MiniBatchKMeans, KMeans


data = pd.read_csv('synthetic_control.csv', sep='\s+', header = None)
data
```

```python
x_train = data
x_train
```

```python
x_train = preprocessing.minmax_scale(x_train)
```

```python
x_train
```

```python
# Compute clustering with Means
k_means = KMeans(init='k-means++', n_clusters=6, n_init=10)
k_means.fit(x_train)
```

```python
# Compute clustering with MiniBatchKMeans
batch_size = 45

mbk = MiniBatchKMeans(init='k-means++', n_clusters=6, batch_size=batch_size,
                      n_init=10, max_no_improvement=10, verbose=0)
mbk.fit(x_train)
```

```python
y_pred_Kmeans = k_means.predict(x_train)
y_pred_Mbk = mbk.predict(x_train)
```
